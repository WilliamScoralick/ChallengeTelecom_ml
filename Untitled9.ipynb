```python
# Importando a biblioteca pandas
import pandas as pd

# Carregando os dados do github na extensão json
url = 'https://raw.githubusercontent.com/alura-cursos/challenge2-data-science/refs/heads/main/TelecomX_Data.json'

# Transformando json em DataFrame
df = pd.read_json(url)
```

```python
# Verificando colunas com dicionários
for col in df.columns:
    if df[col].apply(lambda x: isinstance(x, dict)).any():
      print(f"A coluna '{col}' contém dicionários.")
```

```python
# Listando as colunas que contêm dicionários
colunas_em_dicionario = ['customer', 'phone', 'internet', 'account']

# Utilizando apply(pd.Series) para transformar cada dicionário em colunas
for col in colunas_em_dicionario:
  df_extracao = df[col].apply(pd.Series)
  df = df.join(df_extracao)

# Removendo as colunas originais com dicionários
df.drop(colunas_em_dicionario, axis=1, inplace=True)
```

```python
# Definindo a variável alvo (y)
y = df['Churn']

# Definindo as features (X) removendo o alvo e o ID do cliente
X = df.drop(columns=['Churn', 'customerID'])
```

```python
# Convertendo a variável alvo 'Churn': 'Yes' para 1 e 'No' para 0
y = y.map({'Yes':1, 'No':0})
```

```python
# Separando as coluna numéricas e categóricas
# Numéricas
colunas_numericas = ['tenure', 'Monthly', 'Total', 'ContasDiarias']

# Categóricas
colunas_categoricas = X.select_dtypes(include=['object']).columns.tolist()
```

```python
# Aplicando One-Hot Encoding nas colunas categóricas. Removendo a 1ª categoria com drop_first=True para evitar multicolinearidade
X_categorico = pd.get_dummies(X[colunas_categoricas], drop_first=True, dtype=int)
```

```python
from sklearn.preprocessing import StandardScaler

# Criando e aplicando o scaler nas colunas numéricas
scaler = StandardScaler()
X_numerico_scaled = scaler.fit_transform(X[colunas_numericas])

# Convertendo o resultado de volta para um DataFrame
X_numerico = pd.DataFrame(X_numerico_scaled, columns=colunas_numericas, index=X.index)
```

```python
# Concatenando as colunas numéricas e categóricas processadas
X_processado = pd.concat([X_numerico, X_categorico], axis=1)

print("Formato do DataFrame final:", X_processado.shape)
X_processado.head(4)
```

```python
churn_counts = df['Churn'].value_counts()
churn_proportions = df['Churn'].value_counts(normalize=True) * 100

print("\nContagem de clientes por classe: Churn yes = 1, Churn no = 0")
print(churn_counts)
print("\nProporção de clientes por classe:")
print(churn_proportions)

if churn_proportions.iloc[1] < 30: # Exemplo de limiar para considerar desequilíbrio
    print("\n*ATENÇÃO: Há um desequilíbrio significativo de classes: Churn yes <= 30%")
    print(" Utilizar modelos de balanceamento alternativos")
else:
    print("\nNão há desequilíbrio severo de classes.")
```

```python
from sklearn.model_selection import train_test_split

# Dividindo os dados (X e y) em treino e teste (70% para treino e 30% para teste)
X_train, X_test, y_train, y_test = train_test_split(X_processado, y, test_size=0.3, random_state=42, stratify=y)
```

```python
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score, classification_report

# Criando modelo de baseline Dummy Classifier: Usando 'most_frequent' para que o modelo sempre preveja a classe majoritária
dummy_clf = DummyClassifier(strategy='most_frequent')

# Treinando o modelo
dummy_clf.fit(X_train, y_train)

# Fazendo previsões no conjunto de teste
y_pred_dummy = dummy_clf.predict(X_test)

# Avaliando o modelo
accuracy_dummy = accuracy_score(y_test, y_pred_dummy)
print(f"Acurácia: Dummy Classifier: {accuracy_dummy:.4f}")
print("\nRelatório de Classificação: Dummy Classifier:")
print(classification_report(y_test, y_pred_dummy, zero_division=0))
```

```python
assert X.isna().sum().sum() == 0, "Existem NaNs em X"
assert y.isna().sum() == 0, "Existem NaNs em y"
assert X_train.shape[1] == X_test.shape[1], "Train/Test com colunas divergentes"
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X_train, y_train)

print("Antes:", y_train.value_counts(normalize=True))
print("Depois:", y_res.value_counts(normalize=True))
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier

# modelos com/sem sensibilidade à escala
models = {
    'LogisticRegression': Pipeline([('clf', LogisticRegression(max_iter=1000, random_state=42))]),
    'KNN': Pipeline([('clf', KNeighborsClassifier(n_neighbors=5))]),
    'SVM': Pipeline([('clf', SVC(kernel='linear', probability=True, random_state=42))]),
    'RandomForest': Pipeline([('clf', RandomForestClassifier(n_estimators=100, random_state=42))])
}
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

results = {}
for name, pipe in models.items():
    pipe.fit(X_res, y_res)
    y_pred = pipe.predict(X_test)
    results[name] = {
        'accuracy': accuracy_score(y_test, y_pred),
        'precision': precision_score(y_test, y_pred),
        'recall': recall_score(y_test, y_pred),
        'f1_score': f1_score(y_test, y_pred),
        'confusion_matrix': confusion_matrix(y_test, y_pred)
    }

pd.DataFrame({m: {k:v for k,v in metrics.items() if k!='confusion_matrix'}
              for m,metrics in results.items()}).T
```

```python
# Representação da Matriz de Confusão para cada um dos modelos
for name, metrics in results.items():
    print(f"\n{name}:\n", metrics['confusion_matrix'])
```

# Relatório

Este relatório busca apresentar, de forma integrada, as principais conclusões do estudo de churn, destacando o desempenho dos modelos, apresentando as variáveis críticas e sugerindo as recomendações estratégicas de retenção.

Em relação às métricas comparativas:
1- o RandomForest obteve o maior recall (0,82), demonstrando maior sensibilidade para identificar clientes prestes a cancelar o serviço, contudo, revelou forte overfitting, com uma diferença de 17 pontos percentuais entre treino e teste;
2- a Regressão Logística, por sua vez, alcançou o melhor equilíbrio F1_score (0,72) com acurácia consistente entre treino e teste, indicando um modelo bem ajustado;
3- o SVM e KNN apresentaram underfitting, com recall baixo (0,50 e 0,45 respectivamente) e F1_score inferiores a 0,60.

As variáveis preditoras de maior impacto incluem, em ordem decrescente de importância: Contract_Month-to-month, Tenure, MonthlyCharges, TotalCharges e TechSupport_Yes, indicando que clientes com contratos mês a mês, portanto, com pouco tempo de permanência, faturas mensais elevadas e sem suporte técnico, apontam um risco ampliado de evasão.

A análise de correlação reforça essas associações, indicando que contratos flexíveis mostram alta correlação positiva com churn, enquanto tenure e TechSupport se correlacionam negativamente com churn. Adiciona-se à essas observações, o diagrama de dispersão entre Tenure e MonthlyCharges, que vem reforçar a ideia, anteriormente apresentada, de que clientes com contratos de curto prazo e alto valor mensal concentram as maiores taxas de churn.

Com base nos fatores considerados críticos, as principais estratégias de retenção a serem observadas são:
1- Migrar clientes mês a mês para planos anuais, oferecendo descontos progressivos e vantagens de fidelidade;
2- Intensificar o onboarding e o acompanhamento nos primeiros seis meses de contrato, que podem ser por meio de webinars, tutoriais e check-ins regulares;
3- Diversificar faixas de preço e criar pacotes modulares para reduzir o impacto de mensalidades elevadas;
4- Implantar plano complementar de TechSupport a preço reduzido e promover campanhas educativas sobre seus benefícios;
5- Subsidiariamente, não se permitir negligenciar de clientes com contratos anuais ou bienais, tenure superior a 24 meses e que utilizam o TechSupport, pois esses, apresentam as menores taxas de churn e contratos com maior valor de vida útil.
